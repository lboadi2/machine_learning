```python
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ga_curve = pd.read_csv(os.path.join('algo_results','ga_curves_queen_size_100_iteration_5.csv'))
sa_curve = pd.read_csv(os.path.join('algo_results','sa_curves_queen_size_100_iteration_5.csv'))
rhc_curve = pd.read_csv(os.path.join('algo_results','rhc_curves_queen_size_100_iteration_5.csv'))
mimic_curve = pd.read_csv(os.path.join('algo_results','mimic_curves_queen_size_100_iteration_5.csv'))
```

```python
ga_curve['Mutation Rate'].unique()
```

```python
# Iteration versus fitness
# array([0.25, 0.5 , 0.75])
# array([200, 400, 600], dtype=int64)

fig, axs = plt.subplots()
iteration = ga_curve[(ga_curve['Population Size']==600) & (ga_curve['Mutation Rate']==0.25)]['Iteration']
time = ga_curve[(ga_curve['Population Size']==600) & (ga_curve['Mutation Rate']==0.25)]['FEvals']
axs.plot(iteration, time)
iteration = ga_curve[(ga_curve['Population Size']==600) & (ga_curve['Mutation Rate']==0.5)]['Iteration']
time = ga_curve[(ga_curve['Population Size']==600) & (ga_curve['Mutation Rate']==0.5)]['FEvals']
axs.plot(iteration, time)
iteration = ga_curve[(ga_curve['Population Size']==600) & (ga_curve['Mutation Rate']==0.75)]['Iteration']
time = ga_curve[(ga_curve['Population Size']==600) & (ga_curve['Mutation Rate']==0.75)]['FEvals']
axs.plot(iteration, time)
plt.show()
```
