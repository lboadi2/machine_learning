```python
import matplotlib.pyplot as plt
x = range(0,10)
y = range(0,10)

fig, axs = plt.subplots(nrows=1,ncols=2)
print(axs)
print(type(axs))
```

```python
fig, axs = plt.subplots(nrows=1,ncols=2)
axs[0].plot(x,y)
plt.show()
```

```python
y2 = list(reversed(y))
fig, axs = plt.subplots(nrows=1,ncols=2)
axs[0].plot(x,y)
axs[1].plot(x,y2)
plt.show()
```

```python
fig, axs = plt.subplots(nrows=1,ncols=2)

# first subplot
axs[0].plot(x,y)
axs[0].set_title('positive correlation')

# second subplot
axs[1].plot(x,y2)
axs[1].set_title('negative correlation')

fig.suptitle('two types of perfect correllations')

plt.show()
```

```python
fig, axs = plt.subplots(nrows=2,ncols=1)

axs[0].plot(x,y)
axs[0].set_title("positive correlation")

axs[1].plot(x,y2)
axs[1].set_title('negative correlation')

fig.suptitle('correlation examples')
plt.tight_layout() # removes overlaps between graphs
plt.show()
```

```python
fig, axs = plt.subplots(nrows=2,ncols=3)

axs[0,0].plot(x,y)
axs[0,0].set_title('positive correlation')

axs[0, 1].plot(x,y2)
axs[0, 1].set_title('negative correlation')
plt.tight_layout() # removes overlaps between graphs
plt.show()
```

```python
fig, axs = plt.subplots(nrows=2, ncols=1, figsize=[3,6])
axs[0].plot(x,y)
axs[0].set_title('positive correlation')

axs[1].plot(x,y2)
axs[1].set_title('negative correlation')
fig.suptitle('perfect correlations')
plt.tight_layout()
plt.show()
```
